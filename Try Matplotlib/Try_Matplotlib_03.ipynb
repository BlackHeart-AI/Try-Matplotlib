```python
import matplotlib.pyplot as plt
```

```python
# try pie chart

data = [40, 60]
labels = ["Sixty", "Forty"]

plt.pie(data, labels=labels)

plt.title("Try Pie Chart")
plt.tight_layout()

plt.show()
```

l.png

```python
# add wedgeporps

data = [160, 140]
labels = ["Sixty", "Forty"]

plt.pie(data, labels=labels, wedgeprops={"edgecolor": "#00008B"})

plt.title("Try Pie Chart")
plt.tight_layout()

plt.show()
```

```python
# add more data

data = [30, 20, 10, 40]
labels = ["A", "B", "C", "D"]

plt.pie(data, labels=labels)

plt.title("Try Pie Chart")
plt.tight_layout()

plt.show()
```

```python
# add colors

data = [30, 20, 10, 40]
labels = ["A", "B", "C", "D"]
colors = ["#a3f95d", "#c03354", "#5c2e4a", "#f8345c"]

plt.pie(data, labels=labels, colors=colors)

plt.title("Try Pie Chart")
plt.tight_layout()

plt.show()
```

```python
# Median Developer Salaries by Age
ages = ["25 Age", "26 Age", "27 Age", "28 Age", "29 Age", "30 Age", "31 Age", "32 Age", "33 Age", "34 Age", "35 Age"]
dev_salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]

plt.pie(dev_salary, labels=ages)

plt.title("Median Developer Salaries by Age")

plt.show()
```

```python
# add explode
ages = ["25 Age", "26 Age", "27 Age", "28 Age", "29 Age", "30 Age", "31 Age", "32 Age", "33 Age", "34 Age", "35 Age"]
dev_salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
explode = [0.1, 0, 0, 0, 0]

plt.pie(dev_salary[:5], labels=ages[:5], explode=explode)

plt.title("Median Developer Salaries by Age")

plt.show()
```

```python
# add shadow
ages = ["25 Age", "26 Age", "27 Age", "28 Age", "29 Age", "30 Age", "31 Age", "32 Age", "33 Age", "34 Age", "35 Age"]
dev_salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
explode = [0.1, 0, 0, 0, 0]

plt.pie(dev_salary[:5], labels=ages[:5], explode=explode, shadow=True)

plt.title("Median Developer Salaries by Age")

plt.show()
```
