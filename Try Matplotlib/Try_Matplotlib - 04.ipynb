# Stack Plots

```python
import matplotlib.pyplot as plt
```

```python
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [1, 2, 3, 3, 4, 4, 4, 5, 7]
player2 = [1, 1, 1, 1, 2, 2, 2, 3, 4]
player3 = [1, 2, 3, 4, 5, 1, 6, 1, 1]
```

```python
plt.stackplot(minutes, player1, player2, player3)

plt.show()
```

```python
labels = ["player1", "player2", "player3"]
colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.stackplot(minutes, player1, player2, player3, labels=labels, colors=colors)

plt.title("Stack Plot")
plt.xlabel("Player")
plt.ylabel("Minutes")
plt.legend(loc="upper left") # loc=(0.02, 0.04)

plt.show()
```

```python
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [8, 7, 6, 5, 4, 4, 4, 2, 0]
player2 = [0, 1, 1, 1, 2, 2, 2, 1, 4]
player3 = [0, 2, 3, 4, 3, 1, 3, 1, 1]

labels = ["player1", "player2", "player3"]
colors = ['#6d904f', '#fc4f30', '#008fd5']

plt.stackplot(minutes, player1, player2, player3, labels=labels, colors=colors)

plt.title("Stack Plot")
plt.xlabel("Player")
plt.ylabel("Minutes")
plt.legend(loc="lower left") # loc=(0.02, 0.04)

plt.show()
```
